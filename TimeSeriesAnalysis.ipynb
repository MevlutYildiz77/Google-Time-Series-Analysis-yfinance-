```python
import pandas as pd 
import numpy as np
import yfinance as yf
import datetime
from datetime import date, timedelta

import plotly.graph_objects as go
import plotly.express as px
import plotly.io as pio
pio.templates.default = "plotly_white"
```

```python
d1 = date.today().strftime("%Y-%m-%d")
end_date = d1
```

```python
d2 = date.today() - timedelta(days= 1080)
d2 = d2.strftime("%Y-%m-%d")
start_date = d2
```

```python
df = yf.download('GOOG', 
                 start=start_date,
                 end=end_date,
                 progress=False)
print(df.head())
```

```python
df.tail()
```

```python
df.columns
```

```python
import plotly.express as px
figure = px.line(df, x= df.index,
                 y = "Close",
                 title = "Time Series Anlaysis(Line Plot)")
figure.show()
```

```python
figure = px.line(df, x= df.index,
                 y = "Close",
                 title = "Time Series Anlaysis(Line Plot)")
figure.update_xaxes(rangeslider_visible = True)
figure.show()
```

visualize a candlestick chart to see the trends in the open, high, low, and close prices of Google

```python
import plotly.graph_objects as go
```

```python
figure = go.Figure(data = [go.Candlestick(x=df.index,
                                        open = df["Open"],
                                        high = df["High"],
                                        low = df["Low"],
                                        close = df["Close"])])
figure.update_layout(title = "Time Series Analysis(Candlestick Chart)", 
                     xaxis_rangeslider_visible = False)
figure.show()
```

```python
figure = px.bar(df, x = df.index,
                y = "Close",
title = "Time Series Analysis (Bar Plot)")
figure.show()
```

```python
figure = px.line(df, x = df.index,
                 y = "Close",
                 range_x = ["2021-02-03", "2023-02-03" ],
                 title = "Time Series Analysis(Custom Date Range)")
figure.show()
```

```python
figure = go.Figure(data = [go.Candlestick(x = df.index,
                                          open = df["Open"],
                                          high = df["High"],
                                          low = df["Low"],
                                          close = df["Close"])])
figure.update_layout(title = "Time Series Anlaysis(Candlestick Chart with buttons and slider)")

figure.update_xaxes(
    rangeslider_visible = True,
    rangeselector = dict(
        buttons = list([
            dict(count = 1, label = "1m", step= "month", stepmode = "backward" ),
            dict(count = 6, label = "6m", step = "month", stepmode = "backward"),
            dict(count = 1, label = "YTD", step = "year", stepmode = "todate"),
            dict(count = 1, label = "1y", step = "year", stepmode = "backward" ),
            dict(step = "all")
        ])
    )
)
figure.show()
```

```python
figure = px.line(df, x=df.index, y='Close', 
                 title='Stock Market Analysis with Time Period Selectors')

figure.update_xaxes(
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=3, label="3m", step="month", stepmode="backward"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all")
        ])
    )
)
figure.show()
```

```python
# The weekend or holiday season always affects the stock market. So if you want to remove all the records of the weekend trends from your stock market visualization, below is how you can do it:

figure = px.scatter(df, x = df.index, y="Close", range_x=['2020-04-21', '2023-04-05'],
                    title = "Stock Market Analysis by Hiding Weekend Gaps")

figure.update_xaxes(rangebreaks = [dict(bounds=["sat", "sun"])])
figure.show()
```

```python
figure = px.bar(df, x= df.index, y = "Close", color= "Close")
figure.update_xaxes(rangeslider_visible=True)
figure.show()

```

```python
figure = px.bar(df, x= df.index, y = "Close", color= "Close")
figure.update_xaxes(rangeslider_visible=True)
figure.update_layout(title = "Google Stock Price Analysis", xaxis_rangeslider_visible= False)
figure.update_xaxes(
    rangeselector = dict(
        buttons = list([
            dict(count = 1, label = "1m", step= "month", stepmode = "backward" ),
            dict(count=3, label="3m", step="month", stepmode="backward"),
            dict(count = 6, label = "6m", step = "month", stepmode = "backward"),
            dict(count = 1, label = "YTD", step = "year", stepmode = "todate"),
            dict(count = 1, label = "1y", step = "year", stepmode = "backward" ),
            dict(count=2, label="2y", step="year", stepmode="backward"),
            dict(step = "all")
        ])
    )
)
figure.show()
```

```python
df["Date"] = pd.to_datetime(df.index, format='%Y-%m-%d')
df["Year"] = df["Date"].dt.year
df["Month"]= df["Date"].dt.month
fig = px.line(df, 
              x="Month",
              y="Close",
              color="Year",
              title="Complete timeline of Google")
fig.show()
```
